```python
import matplotlib.pyplot as plt
import random

# Generating random hits for rules
rules = list(range(1, 4001))
hits = [random.randint(1, 15) for _ in range(4000)]

# Generating random 2000 rules to zero out
indices_for_zero_out = random.sample(range(1,4000), 2000)
for idx in indices_for_zero_out:
    hits[idx] = 0

# Generating random rules between 75 and 100 for 3 indices
indices_for_random_hits = random.sample(range(1, 4000),  15)
print(indices_for_random_hits)
for idx in indices_for_random_hits:
    hits[idx] = random.randint(75, 100)

# Threshold line
threshold = [69] * 4000

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(rules, hits, color='blue', alpha=0.7, label='Hits per Rule')
plt.plot(rules, threshold, color='red', linestyle='--', label='Threshold')
plt.xlabel('Rule Number')
plt.ylabel('Hits')
plt.xlim(0,4000)
plt.ylim(0,100)
plt.title('Histogram of Hits per Rule')
plt.legend()
plt.grid(True)
plt.show()
```
